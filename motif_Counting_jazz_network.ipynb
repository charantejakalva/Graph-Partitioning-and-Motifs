```
from graph_tool import clustering as c
import graph_tool.generation as gt
import graph_tool
import numpy as np
```

```
f = open('jazz.txt','r')  
G = np.zeros(shape=(198,198))  
while f: 
  li = f.readline() 
  a,b = li.split()
  print(a,b)
  a = int(a) - 1
  b = int(b) - 1
  G[a][b] = 1 
```

```

motif2 = graph_tool.Graph(directed=True)
adj2 = np.array([[0,1,0],[0,0,0],[1,1,0]])
motif2.add_edge_list(np.transpose(adj2.nonzero()))
motif2 = [motif2]
```

```
print(len(motifs2)) 
print(counts2)
```

```
motif_array2 = []
for i in range(counts2[3]):
    arr = vertexmaps2[3][i].get_array()
    motif_array2.append(arr)
```

```
induced_motif2 = []
non_induced_motif2 = []
ed_count2 = 0;
for k in range(len(motif_array2)):
  ed_count2 = 0;
  for i in range(3):
    for j in range(3):
      a = motif_array2[k][i];
      b = motif_array2[k][j];
      if(G[a][b] == 1):
        ed_count2 = ed_count2 + 1;
  if (ed_count2 == 3):
    induced_motif2.append(motif_array2[k]);
  else: 
    non_induced_motif2.append(motif_array2[k]);

print(len(induced_motif2))
print(len(non_induced_motif2))
```

```
node_counts = {}
for i in range(198):
  node_counts[i] = 0;
for i in range(len(motif_array2)):
  for j in range(3):
    node = motif_array2[i][j];
    node_counts[node] =  node_counts[node] + 1
```
